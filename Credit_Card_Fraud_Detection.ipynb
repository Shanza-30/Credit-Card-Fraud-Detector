```python
# df.drop(['Time', 'Amount'], axis=1, inplace=True) # Commented out as these columns do not exist
```

```python
# Bring scaled columns before the rest (optional)
scaled_cols = ['scaled_amount', 'scaled_time']
other_cols = [col for col in df.columns if col not in scaled_cols + ['Class']]
df = df[scaled_cols + other_cols + ['Class']]
```

```python
# Drop rows with NaN values
df.dropna(inplace=True)

X = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42, k_neighbors=1)
X_train_smote, y_train_smote = sm.fit_resample(X_train, y_train)
```

```python
y_train_smote.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced')
model.fit(X_train, y_train)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Before
sns.countplot(x=y_train)
plt.title("Before SMOTE")
plt.show()

# After
sns.countplot(x=y_train_smote)
plt.title("After SMOTE")
plt.show()
```

```python
# After SMOTE
X_train = X_train_smote
y_train = y_train_smote
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced_subsample'  # Optional if not using SMOTE
)

# Train the model
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
y_pred_proba = rf_model.predict_proba(X_test)[:, 1]  # For AUC
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
confusion_matrix(y_test, y_pred)
classification_report(y_test, y_pred)
("ROC AUC Score:", roc_auc_score(y_test, y_pred_proba))
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(class_weight='balanced', max_iter=1000)
lr_model.fit(X_train, y_train)
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    scale_pos_weight=5,  # Adjust based on class imbalance
    use_label_encoder=False,
    eval_metric='logloss'
)
xgb_model.fit(X_train, y_train)
```
